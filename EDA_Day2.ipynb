## Stacked Barchat/plot

```python
#Stacked barchat/plot
df.groupby(['pclass','sex'])
```

```python
df.groupby(['pclass','sex'])['survived'].mean().unstack().plot(kind='bar',stacked = True)
```

```python
df.groupby(['pclass','who'])['survived'].mean().unstack().plot(kind='bar',stacked = True)
```

```python
df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked = True)

## Average Fare Draw Separate-> child, man, woman 

## df groupby with pclass, who -> pclass in x - axis , -> who in legends(stack) -> Inside the stack -> take the average of fare inside this data(pclass, who) -> after taking average -> unstack -> plot -> how plot -> average inside the bar plot and stack it again 
```

```python
## how to change position of xticks in matplotlib

df.groupby(['pclass','who'])['fare'].mean().unstack().plot(kind='bar',stacked = True)
plt.xticks(rotation=0)   # 0, 45, 90 etc 
```
