# Data Wrangling

Data Wrangling is the process of `cleaning`, `transforming` and `organizing` data so that it can be used for `analysis and visualization`. It is an important step in the data analysis process because raw data is often `incomplete`, `inconsistent`, and in an `unstructured format`, which can make it difficult to work with. 
- Data Wrangling helps to make the data more `consistent`, `accurate`, and `useful` for analysis and decision making.

- Data Science -> Data Driven Decision Making -> Taking Decision on the basis of data -> called data science

- Data Wrangling -> making decisions useful, consistent and accurate

Data Wrangling is a special type of data science, in which we clean the data, do pre-processing so that the data becomes ready for further analysis.

The Data is never True. we have to clean the data, clean it so much that there is no chance of any error in it.

And if any of analysis gives these hypotheses, we make some guess, then the chances of that guess proving wrong in the future are reduced.

#### 4.2.1.2 IQR Method

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)

IQR = Q3 - Q1
print(IQR)
```

Make an equation -> which we have use to remove outlier with IQR level

- Calculate IQR

IQR = Q3 − Q1

- Lower Bound 

Lower Bound = Q1 − 1.5 * IQR	​


- Upper Bound

Upper Bound = Q3 + 1.5 * IQR


- Outlier Rule

A value is an outlier if: x < Q1 − 1.5 * IQR or x > Q3 + 1.5 * IQR
	​


```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['age'] > lower_bound) & (df['age'] < upper_bound)]
```

```python
df.shape

## 891 - 825 = 66 -> Outlier remove -> age
```

```python
## For Fare Variable

Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)

IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['fare'] > lower_bound) & (df['fare'] < upper_bound)]
```

#### 4.2.1.3 Z-Score Method

The Z-score method detects outliers by measuring how far a data point is from the mean, in terms of standard deviations.

- Use library -> scipy -> for use stats term

```python
%pip install scipy
```

```python
from scipy import stats
import numpy as np

zscore = np.abs(stats.zscore(df['age']))
threshold = 3
df = df[zscore < threshold]
```

```python
df.shape
```

```python
zscore = np.abs(stats.zscore(df[['age', 'fare']]))
threshold = 3

df = df[(zscore < threshold).all(axis=1)]
```

#### 5.1.2 Standard Scaler

- No fixed range (approximately −∞ to +∞)

- Standard Scaler is a normalization technique that transforms data so that it has a mean of 0 and a standard deviation of 1. It is commonly used for machine learning algorithms that assume normally distributed data.

Only Change Scaling Unit/Method

#### Formula 

x_stand = (x - mean(x)) / std(x)

Where:

- x_stand is the standardized variable
- x is the original variable
- mean(X) is the mean of the variable x
- std(x) is the standard deviation of variable x

```python
# Import Libraries
import pandas as pd 
from sklearn.preprocessing import StandardScaler

# data 
df

# Select Columns to Normalize
cols_to_noramlize = ['age', 'fare']

# Create the Scalar function/tool
scaler = StandardScaler()

# Fit and Transform the data on scaler or vice versa 
df[cols_to_noramlize] = scaler.fit_transform(df[cols_to_noramlize])
```
